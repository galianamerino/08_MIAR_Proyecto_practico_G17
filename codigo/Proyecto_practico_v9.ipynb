**Definición de la clase AtariProcessor**

````python
class AtariProcessor(Processor):
    def process_observation(self, observation):
        assert observation.ndim == 3  # (height, width, channel)
        img = Image.fromarray(observation)
        #img = img.resize(INPUT_SHAPE).convert('L')
        #img = img.resize((INPUT_SHAPE2[1], INPUT_SHAPE2[0])).convert('L')
        img = img.resize((INPUT_SHAPE[1], INPUT_SHAPE[0])).convert('L')
        processed_observation = np.array(img)
        assert processed_observation.shape == INPUT_SHAPE
        return processed_observation.astype('uint8')


    def process_state_batch(self, batch):
        processed_batch = batch.astype('float32') / 255.
        return processed_batch

    def process_reward(self, reward):
        return np.clip(reward, -1., 1.)
````

###3.4. Configuración de clases adicionales

**Selección o restricción de acciones**

````python
# Definir la clase que permite restringir o seleccionar las posibles
# acciones del entorno

class SeleccionAcciones(gym.ActionWrapper):
    def __init__(self, env, allowed_actions):
        super(SeleccionAcciones, self).__init__(env)
        self.allowed_actions = allowed_actions
        self.action_space = gym.spaces.Discrete(len(allowed_actions))

    def action(self, act):
        return self.allowed_actions[act]

    def reverse_action(self, act):
        return self.allowed_actions.index(act)
````

````python
# Aplicación de la función al entorno

env = SeleccionAcciones(env, hiperparametros['Acciones_permitidas'])

# Confirmar tamaño del espacio de acción reducido
nb_actions = env.action_space.n
print("Acciones utilizadas:", nb_actions)
````

###3.5. Implementación de la red neuronal

````python

# Dimensiones de las observaciones de entrada
input_shape = (WINDOW_LENGTH,) + INPUT_SHAPE

# Definición de la red neuronal
model = Sequential()

model.add(Permute((2, 3, 1), input_shape=input_shape))

model.add(Convolution2D(32, (8, 8), strides=(4, 4)))
model.add(Activation('relu'))
model.add(Convolution2D(64, (4, 4), strides=(2, 2)))
model.add(Activation('relu'))
model.add(Convolution2D(64, (3, 3), strides=(1, 1)))
model.add(Activation('relu'))
model.add(Flatten())
model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dense(nb_actions))
model.add(Activation('linear'))
print(model.summary())
````

###3.6. Implementación de la solución DQN

**Inicialización del procesado de las observaciones y de la memoria del entrenamiento**

````python
# Define la memoria de entrenamiento
memory = SequentialMemory(limit=hiperparametros['REPLAY_MEMORY'], window_length=hiperparametros['WINDOW_LENGTH'])

# Crea el objeto que procesa las observaciones, 'processor'
processor = AtariProcessor()
````

**Estrategia de exploración**

````python
# Define la estrategia de exploración.

# Las posibles estrategias implementadas en keras son:

# LinearAnnealedPolicy: Es una política ε-Greedy donde ε disminuye linealmente a lo largo del tiempo.
# SoftmaxPolicy Selecciona acciones basadas en una distribución de probabilidad generada por una función softmax aplicada a los valores Q.
# EpsGreedyQPolicy Selecciona la acción con el valor Q más alto con probabilidad 1-ε y una acción aleatoria con probabilidad ε.
# GreedyQPolicy Siempre selecciona la acción con el valor Q más alto.
# BoltzmannQPolicy Selecciona acciones basadas en una distribución de probabilidad generada por la función Boltzmann aplicada a los valores Q.
# MaxBoltzmannQPolicy Combina Boltzmann y ε-Greedy, seleccionando con probabilidad 1-ε la acción Boltzmann y con probabilidad ε una acción aleatoria.
# BoltzmannGumbelQPolicy Utiliza la distribución de Gumbel para generar una política Boltzmann mejorada.

# En la solución propuesta, utilizaremos al estrategia LinearAnnealedPolicy.
# Esta estrategia comienza explorando al 100% (ε = 1.0) y va reduciendo hasta el mínimo (ε = 0.1 o similar).
# Por tanto, permite que el agente explore mucho al inicio y se vuelva más "inteligente" con el tiempo.


policy = LinearAnnealedPolicy(
    EpsGreedyQPolicy(),       # Estrategia epsilon-greedy
    attr='eps',               # Atributo que se va a reducir con el tiempo: ε
    value_max=1.,             # Valor inicial de ε → el agente explora el 100% del tiempo al inicio
    value_min=0.01,            # Valor final de ε tras nb_steps → el agente aún explora un 10% del tiempo
    value_test=0.01,          # Valor de ε durante la fase de evaluación (test), sin entrenamiento
    nb_steps=100000            # Número de pasos durante los cuales se reduce ε de 1.0 a 0.1 linealmente
                              # Este valor,60000, se ha escogido teniendo en cuenta el número de pasos de entrenamiento
                              # en cada ciclo, que en nuestro caso es de 100000.
)
````

**Incialización del agente DQN**

````python
# Inicialización del agente DQN

dqn = DQNAgent(
    model=model,
    nb_actions=nb_actions,
    policy=policy,
    memory=memory,
    processor=processor,
    nb_steps_warmup = NB_STEPS_WARMUP,
    gamma=hiperparametros['GAMMA'],
    target_model_update = hiperparametros['TARGET_MODEL_UPDATE'],
    train_interval = hiperparametros['TRAIN_INTERVAL'],
    delta_clip = DELTA_CLIP
)

# Optimizador Adam con una tasa de aprendizaje definida por la variable LEARNING_RATE
optimizer = Adam(learning_rate = hiperparametros['LEARNING_RATE'])

# Configuración del agente
dqn.compile(optimizer=optimizer, metrics=['mae'])
````

**Callback para guardar el mejor modelo del entrenamiento**

````python
# Callback para guardar mejor modelo por ciclo y global
class GuardarMejorModeloEntrenamiento(Callback):
    def __init__(self, path_modelo_ciclo, path_modelo_global=None, global_tracker=None):
        self.mejor_recompensa = float('-inf')
        self.path_modelo_ciclo = path_modelo_ciclo
        self.path_modelo_global = path_modelo_global
        self.global_tracker = global_tracker

    def on_episode_end(self, episode, logs={}):
        recompensa = logs.get('episode_reward', 0)

        # Mejor del ciclo
        if recompensa > self.mejor_recompensa:
            self.mejor_recompensa = recompensa
            self.model.save_weights(self.path_modelo_ciclo, overwrite=True)
            print(f"Guardado mejor modelo del ciclo: {self.path_modelo_ciclo} (recompensa: {recompensa:.2f})")

        # Mejor global
        if self.global_tracker and recompensa > self.global_tracker['recompensa']:
            self.global_tracker['recompensa'] = recompensa
            self.model.save_weights(self.path_modelo_global, overwrite=True)
            print(f"Guardado nuevo mejor modelo GLOBAL: {self.path_modelo_global} (recompensa: {recompensa:.2f})")
````

**Callback para parar entrenamiento cuando se supera el umbral requerido durante N episodios seguidos**

````python
class EarlyStoppingEntrenamiento(Callback):
    def __init__(self, umbral=20, episodios_consecutivos=10, ruta_modelo='mejor_modelo_entrenamiento.h5f', stop_flag=None):
        super().__init__()
        self.umbral = umbral
        self.episodios_consecutivos = episodios_consecutivos
        self.contador = 0
        self.ruta_modelo = ruta_modelo
        self.stop_flag = stop_flag  # ← referencia externa

    def on_episode_end(self, episode, logs={}):
        recompensa = logs.get('episode_reward', 0)

        if recompensa >= self.umbral:
            self.contador += 1
            #print(f"Episodio {episode}: recompensa {recompensa:.2f} ≥ {self.umbral} ({self.contador}/{self.episodios_consecutivos})")

            if self.contador >= self.episodios_consecutivos:
                print("Parada temprana alcanzada. Guardando modelo y deteniendo entrenamiento.")
                self.model.save_weights(self.ruta_modelo, overwrite=True)
                self.model.stop_training = True
                if self.stop_flag is not None:
                    self.stop_flag["detener"] = True  # ← señal para el bucle externo
        else:
            #print(f"Episodio {episode}: recompensa {recompensa:.2f} < {self.umbral}. Reiniciando contador.")
            self.contador = 0
````

**Callback para mostrar la evolución del parámetro epsilon del policy**

````python
from rl.callbacks import Callback

class EpsilonLogger(Callback):
    def __init__(self, intervalo=10000):
        self.intervalo = intervalo
        self.total_steps = 0

    def on_step_end(self, step, logs=None):
        self.total_steps += 1

        if self.total_steps % self.intervalo == 0:
            # Calcula el valor de epsilon manualmente, imitando LinearAnnealedPolicy
            policy = self.model.policy
            t = min(1.0, self.total_steps / float(policy.nb_steps))
            current_eps = policy.value_max - t * (policy.value_max - policy.value_min)
            print(f"Paso {self.total_steps}: ε = {current_eps:.4f}")


epsilon_logger = EpsilonLogger(intervalo=10000)
````

**Entrenamiento**

Bucle de entrenamiento con criterio de parada inteligente.



1) Entrenamiento en ciclos hasta que se cumpla:
* Recompensa igual o superior al umbral durante T episodios seguidos, *episodios_umbral*
* Recompensa promedio igual o superior al umbral en el test realizado al final de cada ciclo.
* Se llega al número máximo de ciclos, *max_ciclos*

2) En caso de interrupción, posibilidad de continuar desde:

* El modelo final del ciclo anterior, o

* El mejor modelo del ciclo anterior.

3) Callbacks personalizados para:

* Guardar el mejor modelo del ciclo.

* Guardar el mejor modelo global.

* Guardar logs .json con FileLogger.

* Guardar checkpoints periódicos con ModelIntervalCheckpoint.

* Guardar el modelo y terminar cuando la recompensa sea igual o superior al umbral durante T episodios seguidos.

4) Un archivo entrenamiento.json con resumen completo del proceso.

````python
 # Configuración general
max_ciclos = 20
umbral = 20
episodios_umbral=10
repeticiones_deseadas = 10
mejorado_dos_veces = 0  # Se utilizarían en caso que repeticiones_deseadas fuera mayor que 1
global_tracker = {'recompensa': float('-inf')}
historial_entrenamiento = []

# El entrenamiento se ha configurado con un número de pasos límite igual a:
# max_ciclos * pasos_final
# En este caso, los pasos por ciclo serían 100 000 y el límite posible de pasos se ha fijado en 2 000 000 pasos
pasos_intervalo=20000
pasos_final=10*pasos_intervalo
pasos_check=5*pasos_intervalo

pasos_acumulados=0

# En caso de interrupción, se puede continuar desde el último ciclo entrenado
# utilizando el último modelo del ciclo (opción 0) o el mejor modelo del ciclo
# (opción 1).
cargar_para_entrenamiento = 0  # Por defecto, se deja seguir desde el modelo final (opción 0)

# Comprueba que existan los directorios
ruta_pesos=ruta + "pesos"
os.makedirs(ruta_pesos, exist_ok=True)

os.makedirs(ruta_pesos + "/mejores", exist_ok=True)
os.makedirs(ruta_pesos + "/logs", exist_ok=True)
os.makedirs(ruta_pesos + "/checkpoints", exist_ok=True)
os.makedirs(ruta_pesos + "/ciclo", exist_ok=True)

# Busqueda de modelos finales guardados de un entrenamiento previo
# Permite continuar el entrenamiento después de una interrupción
modelos_guardados = []
pattern = re.compile(r'modelo_final_ciclo_(\d+)\.h5f.index')
carpeta_modelos = os.path.join(ruta, "pesos/ciclo")

if os.path.exists(carpeta_modelos):
    for archivo in os.listdir(carpeta_modelos):
        match = pattern.match(archivo)
        if match:
            modelos_guardados.append(int(match.group(1)))

modelos_guardados = sorted(modelos_guardados)

# Opción de continuar desde el último modelo guardado o reiniciar el entrenamiento desde inicio
if modelos_guardados:
    print("Modelos guardados encontrados:")
    for ciclo in modelos_guardados:
        print(f"  - Ciclo {ciclo}")
    try:
        valor_inicial = input(f"¿Desde qué ciclo quieres continuar? (inicio = 1 / último disponible = {modelos_guardados[-1] + 1}): ")
        ciclo_inicial = int(valor_inicial)
    except ValueError:
        ciclo_inicial = modelos_guardados[-1] + 1
        print(f"Entrada inválida. Continuando desde el ciclo {ciclo_inicial}.")

else:
    print("No se encontraron modelos previos. Iniciando desde el ciclo 1.")
    ciclo_inicial = 1


###
# 🔁 Verifica si hay que limpiar ciclos posteriores al seleccionado
if modelos_guardados and ciclo_inicial <= modelos_guardados[-1]:
   print(f"🧹 Eliminando memoria, policy, pesos y logs desde el ciclo {ciclo_inicial} hasta {modelos_guardados[-1]}")

   for ciclo_borrar in range(ciclo_inicial, modelos_guardados[-1] + 1):
       archivo_memoria = ruta + f"pesos/memoria_ciclo_{ciclo_borrar}.pkl"
       archivo_policy = ruta + f"pesos/estado_policy_ciclo_{ciclo_borrar}.json"
       archivo_pesos = ruta + f"pesos/ciclo/modelo_final_ciclo_{ciclo_borrar}.h5f"
       archivo_pesos_index = archivo_pesos + ".index"
       archivo_log = ruta + f"pesos/logs/entrenamiento_ciclo_{ciclo_borrar}.json"

       for archivo in [archivo_memoria, archivo_policy, archivo_pesos, archivo_pesos_index, archivo_log]:
           if os.path.exists(archivo):
               os.remove(archivo)
               print(f"   ⛔ Eliminado: {archivo}")

####

if ciclo_inicial>1:
    # Continua entrenamiento con el mejor modelo
    if cargar_para_entrenamiento == 1:
        path_cargar_entrenamiento = ruta + "pesos/mejores/mejor_modelo_global.h5f"

    # Continua entrenamiento con el último modelo
    else:
        path_cargar_entrenamiento = ruta + f"pesos/ciclo/modelo_final_ciclo_{ciclo_inicial-1}.h5f"


    dqn.load_weights(path_cargar_entrenamiento)


# Indicador de parada temprana
stop_flag = {"detener": False}


# Rutas de estado
estado_policy_path = ruta + "pesos/estado_policy.json"
ruta_memoria = ruta + "pesos/memoria.pkl"

if ciclo_inicial > 1:
    ciclo_previo = ciclo_inicial - 1

    # Restaurar estado del policy (por ciclo si está disponible)
    estado_policy_ciclo_path = ruta + f"pesos/estado_policy_ciclo_{ciclo_previo}.json"
    try:
        with open(estado_policy_ciclo_path, "r") as f:
            estado = json.load(f)
            dqn.step = estado.get("steps", 0)
            pasos_acumulados=dqn.step
            print(f"✔️ Restaurado dqn.policy.nb_steps desde ciclo {ciclo_previo}: {dqn.step}")

    except FileNotFoundError:
        print(f"⚠️ No se encontró el archivo de estado del policy del ciclo {ciclo_previo}. Se usará steps = 0.")

####

# Bucle principal de entrenamiento
for ciclo in range(ciclo_inicial, max_ciclos + 1):
    print(f"\nCICLO {ciclo}/{max_ciclos}")

    # Rutas para este ciclo
    path_modelo_final = ruta + f"pesos/ciclo/modelo_final_ciclo_{ciclo}.h5f"
    path_modelo_ciclo = ruta + f"pesos/mejores/mejor_modelo_entrenamiento_ciclo_{ciclo}.h5f"
    path_modelo_ciclo_early_stop=ruta + f"/pesos/mejores/mejor_modelo_early_stop_ciclo_{ciclo}.h5f"
    path_modelo_global = ruta + "pesos/mejores/mejor_modelo_global.h5f"
    path_log = ruta + f"pesos/logs/entrenamiento_ciclo_{ciclo}.json"
    path_checkpoints = ruta + f"pesos/checkpoints/modelo_ciclo_{ciclo}_step_{{step}}.h5f"


    # Callbacks para el entrenamiento

    callbacks_ciclo = [
        GuardarMejorModeloEntrenamiento(
            path_modelo_ciclo=path_modelo_ciclo,
            path_modelo_global=path_modelo_global,
            global_tracker=global_tracker
        ),
        EarlyStoppingEntrenamiento(
            umbral=umbral,
            episodios_consecutivos=episodios_umbral,
            ruta_modelo=path_modelo_ciclo_early_stop,
            stop_flag=stop_flag
        ),
        FileLogger(path_log, interval=100),
        ModelIntervalCheckpoint(path_checkpoints, interval=pasos_check),
        epsilon_logger
    ]


    # ENTRENAMIENTO

    # Solo para el primer ciclo, mantener warmup
    if ciclo == 1: #ciclo_inicial:
        dqn.nb_steps_warmup = NB_STEPS_WARMUP

    else:
        if ciclo == ciclo_inicial:
           dqn.nb_steps_warmup=pasos_intervalo
        else:
          dqn.nb_steps_warmup = 0

        dqn.policy.value_max=dqn.policy.value_min

    dqn.step=pasos_acumulados

    dqn.fit(
        env,
        callbacks=callbacks_ciclo,
        nb_steps=pasos_final,
        log_interval=pasos_intervalo,
        visualize=False
    )


    # Guardar modelo final del ciclo
    dqn.save_weights(path_modelo_final, overwrite=True)


    # TEST
    resultado = dqn.test(env, nb_episodes=10, visualize=False)
    recompensas = resultado.history['episode_reward']
    promedio = sum(recompensas) / len(recompensas)
    print(f"Recompensa promedio en test: {promedio:.2f}")


    fue_mejor_global = (global_tracker['recompensa'] == max(recompensas))

    # Guardar resumen de ciclo
    historial_entrenamiento.append({
      "ciclo": ciclo,
      "recompensa_promedio_test": promedio,
      "supero_umbral": promedio >= umbral,
      "modelo_final": path_modelo_final,
      "mejor_modelo_ciclo": path_modelo_ciclo,
      "mejor_modelo_global": path_modelo_global if fue_mejor_global else None,
      "log_entrenamiento": path_log,
      "cargar_para_entrenamiento": "mejor" if cargar_para_entrenamiento else "final"

    })


    # Guardar estado del policy (steps de LinearAnnealedPolicy)
    with open(estado_policy_path, "w") as f:
        #pasos_acumulados += dqn.step
        pasos_acumulados += pasos_final
        json.dump({"steps": pasos_acumulados}, f)
        print(f"💾 Estado del policy guardado: steps = {pasos_acumulados}")


    # Guardar una copia del estado del policy con nombre del ciclo
    estado_policy_ciclo_path = ruta + f"pesos/estado_policy_ciclo_{ciclo}.json"
    with open(estado_policy_ciclo_path, "w") as f:
        #json.dump({"steps": dqn.step}, f)
        json.dump({"steps":pasos_acumulados}, f)


    # Criterio de parada por test
    if promedio >= umbral:
      mejorado_dos_veces += 1
      print(f"Mejora {mejorado_dos_veces}/{repeticiones_deseadas}")
    else:
      mejorado_dos_veces = 0
      print("No alcanzó el umbral. Reiniciando contador.")

    if mejorado_dos_veces >= repeticiones_deseadas:
      print("\nEntrenamiento finalizado exitosamente.")
      break
    else:
      print("\nSe alcanzó el número máximo de ciclos sin cumplir el criterio.")


    # Criterio de parada temprana
    if stop_flag["detener"]:
        print("Entrenamiento detenido por condición de parada temprana.")
        break



# Función para convertir datos a tipo json
def convertir_tipo_json(obj):
    if isinstance(obj, (np.integer, np.int_, np.int32, np.int64)):
        return int(obj)
    elif isinstance(obj, (np.floating, np.float_, np.float32, np.float64)):
        return float(obj)
    elif isinstance(obj, (np.bool_, bool)):
        return bool(obj)
    elif isinstance(obj, np.ndarray):
        return obj.tolist()
    else:
        raise TypeError(f"Tipo no serializable: {type(obj)}")

# Guardar como JSON con conversión
with open(ruta + "pesos/entrenamiento.json", "w") as f:
    json.dump(historial_entrenamiento, f, indent=4, default=convertir_tipo_json)

print("Historial completo guardado en: pesos/entrenamiento.json")
````

**Función para representar gráficamente los resultados obtenidos para cada uno de los episodios ejecutados**

````python


def cargar_datos_experimento(log_dir,metricas):

    """Carga y concatena los datos de todos los ficheros y elimina los episodios con NaN en cualquier métrica."""
    data_concatenada = {clave: [] for clave in metricas}

    #ficheros = sorted(glob(os.path.join(log_dir, "entrenamiento_ciclo_*.json")))



    ficheros = glob(os.path.join(log_dir, "entrenamiento_ciclo_*.json"))
    ficheros = sorted(ficheros, key=lambda f: int(re.search(r"entrenamiento_ciclo_(\d+)\.json", f).group(1)))

    for fichero in ficheros:
        with open(fichero, 'r') as f:
            datos = json.load(f)
            for clave in metricas:
                data_concatenada[clave].extend(datos[clave])

    # Convierte a arrays de NumPy
    datos_np = {clave: np.array(valores) for clave, valores in data_concatenada.items()}

    # Identifica los episodios válidos (sin NaNs en ninguna métrica)
    indices_validos = ~np.isnan(np.column_stack([datos_np[clave] for clave in metricas])).any(axis=1)

    # Filtra los episodios inválidos
    datos_filtrados = {clave: valores[indices_validos] for clave, valores in datos_np.items()}

    return datos_filtrados


def representar_experimentos(experiments,metricas):

    fig, axs = plt.subplots(2, 3, figsize=(20, 10))
    axs = axs.flatten()

    for i, metrica in enumerate(metricas):

        for nombre, ruta in experiments.items():
            datos = cargar_datos_experimento(ruta,metricas)
            valores = np.array(datos[metrica])
            episodios = np.arange(len(valores))
            # Filtrar NaNs
            #episodios = episodios[~np.isnan(valores)]
            #valores = valores[~np.isnan(valores)]
            axs[i].plot(episodios, valores, label=nombre)

        axs[i].set_title(metrica)
        axs[i].set_xlabel("Episodios")
        axs[i].set_ylabel(metrica)
        axs[i].legend()
        axs[i].grid(True)

    plt.tight_layout()
    plt.show()
````

**Función para obtener y representar la recompensa promedio junto con su correspondiente desviación estandar**

````python


def representar_recompensa_promedio_y_varianza(experiments, metrica='episode_reward', tam_bloque=50):
    """
    Representa la evolución de la recompensa promediada en bloques, con bandas de desviación estándar.
    Además, devuelve los datos en un diccionario de DataFrames.

    Parámetros:
    - experiments: dict con nombre del experimento y ruta al log_dir
    - metrica: métrica a representar (por defecto: 'episode_reward')
    - tam_bloque: número de episodios por bloque para promediar y calcular desviación

    Devuelve:
    - dict: {nombre_experimento: DataFrame con columnas ['episodio_centro', 'media', 'min', 'max']}
    """
    plt.figure(figsize=(14, 6))
    resultados = {}

    for nombre, ruta in experiments.items():
        datos = cargar_datos_experimento(ruta, [metrica])
        recompensas = np.array(datos[metrica])

        num_bloques = len(recompensas) // tam_bloque

        medias, desviaciones, episodios_centro = [], [], []

        for i in range(num_bloques):
            inicio = i * tam_bloque
            fin = inicio + tam_bloque
            bloque = recompensas[inicio:fin]

            media = np.mean(bloque)
            std = np.std(bloque)

            medias.append(media)
            desviaciones.append(std)
            episodios_centro.append(inicio + tam_bloque // 2)

        medias = np.array(medias)
        desviaciones = np.array(desviaciones)
        episodios_centro = np.array(episodios_centro)
        min_band = medias - desviaciones
        max_band = medias + desviaciones

        # Guardar DataFrame para análisis posterior
        df_resultado = pd.DataFrame({
            "episodio_centro": episodios_centro,
            "media": medias,
            "min": min_band,
            "max": max_band
        })
        resultados[nombre] = df_resultado

        # Gráfica
        plt.plot(episodios_centro, medias, label=nombre)
        plt.fill_between(episodios_centro, min_band, max_band, alpha=0.2)

    plt.title(f"Recompensa promedio por bloques de {tam_bloque} episodios ± 1σ")
    plt.xlabel("Episodio")
    plt.ylabel("Recompensa promedio")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    return resultados
````

**Función para representar gráficamente los resultados promedio para cada uno de los periodos ejecutados en cada ciclo**

````python
def load_and_average_experiment(log_folder, fields_to_plot):

    #log_files = sorted(glob.glob(os.path.join(log_folder, "entrenamiento_ciclo_*.json")))

    log_files = glob(os.path.join(log_folder, "entrenamiento_ciclo_*.json"))
    log_files = sorted(log_files, key=lambda f: int(re.search(r"entrenamiento_ciclo_(\d+)\.json", f).group(1)))


    dfs = []

    for file in log_files:
        with open(file, 'r') as f:
            raw_data = json.load(f)

        if isinstance(raw_data, dict):
            min_len = min(len(v) for v in raw_data.values())
            trimmed = {k: v[:min_len] for k, v in raw_data.items()}
            records = [dict(zip(trimmed.keys(), values)) for values in zip(*trimmed.values())]
            df = pd.DataFrame(records)
        else:
            df = pd.DataFrame(raw_data)

        dfs.append(df)

    # Alinea los ficheros y los promedia
    combined_df = pd.concat(dfs).groupby('episode').mean(numeric_only=True).reset_index()
    return combined_df[["episode"] + fields_to_plot]


def plot_avg_experiments(experiments, fields_to_plot):
    import matplotlib.pyplot as plt
    import numpy as np

    n_fields = len(fields_to_plot)
    fig, axs = plt.subplots(2,3, figsize=(15, 10))
    axs = axs.flatten()

    if n_fields == 1:
        axs = [axs]

    colors = plt.cm.viridis(np.linspace(0, 1, len(experiments)))

    for (label, folder), color in zip(experiments.items(), colors):
        df = load_and_average_experiment(folder, fields_to_plot)
        for i, field in enumerate(fields_to_plot):
            axs[i].plot(df["episode"], df[field], label=label, color=color)
            axs[i].set_xlabel("Episodios")
            axs[i].set_ylabel(field)
            axs[i].set_title(field)
            axs[i].grid(True)

    for ax in axs:
        ax.legend()

    plt.tight_layout()
    plt.show()
````

````python


def comparar_hiperparametros(experiments_test):
    """
    Lee archivos 'hiperparametros.csv' desde varios experimentos y
    muestra una tabla comparativa bonita usando display().

    Oculta algunos parámetros que no se desean mostrar.
    """
    tabla_final = pd.DataFrame()

    for _, ruta_experimento in experiments_test.items():
        nombre_carpeta = os.path.basename(ruta_experimento)
        nombre_experimento = nombre_carpeta.replace("_", " ").capitalize()

        ruta_csv = os.path.join(ruta_experimento, "hiperparametros.csv")

        if not os.path.exists(ruta_csv):
            print(f"⚠️ Archivo no encontrado: {ruta_csv}")
            continue

        try:
            df = pd.read_csv(ruta_csv, sep=None, engine='python')
            df.columns = ['Nombre', 'Valor']
            df.set_index('Nombre', inplace=True)

            tabla_final[nombre_experimento] = df['Valor']
        except Exception as e:
            print(f"❌ Error al procesar {ruta_csv}: {e}")

    # Transponer: experimentos como filas
    tabla_final = tabla_final.T

    # ❌ Parámetros que no se quieren mostrar
    parametros_a_ocultar = ["INPUT_SHAPE", "NB_STEPS_WARMUP", "REPLAY_MEMORY","BATCH_SIZE","DELTA_CLIP"]
    tabla_final = tabla_final.drop(columns=parametros_a_ocultar, errors='ignore')

    # 👁️ Mostrar de forma bonita
    display(tabla_final)

    return tabla_final

````

**Función para representar en una tabla los resultados obtenidos entre varios experimentos**

````python
def resumen_test_multiple_desde_csv(experiments):
    resultados = []


    for nombre, ruta in experiments.items():
        # Formamos el nombre del archivo en el formato adecuado

        match = re.search(r"experimento_(\d+)", ruta)
        tipo_exp = f"EXPERIMENTO_{match.group(1)}"


        nombre_csv = f"Resultados_{tipo_exp}.csv"
        path_csv = os.path.join(ruta, nombre_csv)

        if not os.path.isfile(path_csv):
            print(f"Archivo no encontrado: {path_csv}")
            continue

        df = pd.read_csv(path_csv)
        row = df.iloc[0]

        resultados.append({
            "Experimento": tipo_exp,
            "Recompensa Mínima": row["Recompensa Mínima"],
            "Recompensa Máxima": row["Recompensa Máxima"],
            "Recompensa Promedio": row["Recompensa Promedio"]
        })

    return pd.DataFrame(resultados)
````

**Resultados obtenidos en el entrenamiento para cada uno de los episodios ejecutados**

A continuación se comparan los diferentes experimentos realizados en función de los siguientes parámetros medidos durante el entrenamiento.



1) **loss** (pérdida). Se trata de la función de pérdida utilizada para actualizar la red. En DQN, típicamente es el error cuadrático medio entre el valor Q estimado y el valor objetivo (target Q-value).
Al principio, puede ser aleatorio, porque el agente no sabe qué hacer.  Luego de ello, la loss se estabiliceen torno a un valor con oscilaciones pequeñas, sin  explosiones/NaNs, mientras la recompensa media y mean_q se mantienen o crecen lentamente.

2) **mae** (Mean Absolute Error). Es el error absoluto medio entre los Q predichos y sus objetivos. Su comportamiento esperado es muy parecido al de la loss: alto y ruidoso al inicio, luego se estabiliza en un rango acotado. Como no eleva el error al cuadrado, es menos sensible a outliers, así que suele verse más suave que la loss.

3) **mean_q** (Valor Q medio). Es el valor promedio de las estimaciones de Q del modelo para las acciones seleccionadas. Da una idea de lo confiado que está el agente en sus decisiones. Valores bajos podrían significar que el agente espera recompensas pobres. Aumenta si el agente encuentra una buena política que da buenas recompensas. Al principio puede ser bajo, pero debería aumentar progresivamente si el agente aprende una política efectiva.

4) **episode_reward** (puntuación). Suma de recompensas de cada episodio. Es la métrica principal de desempeño, al inicio tiene mucha varianza, pero lo deseable es que muestre una tendencia ascendente y se estabilice en niveles altos.

5) **nb_episode_steps** (número de pasos por episodio). Se refiere a cuántas acciones (pasos del agente) se ejecutan dentro de un solo episodio de entrenamiento o evaluación. Valores altos  indican que el agente sobrevive o se desempeña bien durante más tiempo, mientras que valores bajos (sobre 20) indican que el agente falla pronto o toma malas decisiones que terminan el episodio rápidamente.

6) **duration** (duración). Este parámetro nos indica la duración por episodio, aunque está depende del entorno de ejecución disponible en cada momento del entrenamiento.
